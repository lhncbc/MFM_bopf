### Use "out_flag" variable to control whether files are written or not

```python
out_flag = False

from datetime import datetime
datestr = datetime.today().strftime('%Y-%m-%d')
```

```python
# Write out complete dataset, including Sitenum
filename = 'MFM_CSL_d6_tl_PI_complete_' + datestr + '.csv'
linkname = 'CSL_tl_PI_comp.csv'
if out_flag:
    tl_PI_mod.to_csv(filename, header=True)
    if os.path.exists(linkname):
        os.remove(linkname)
    os.symlink(filename, linkname)
else:
    print(f'File not written: {filename}')
```

```python
import pandas as pd
import os
import glob

pd.options.display.max_columns = None
pd.options.display.max_rows = None

#os.chdir('/MFMDatasets/MFM_bopf/src/multi_predict/NB/output')
os.chdir('/MFMDatasets/MFM_bopf/src/multi_predict/LR/output')
#os.chdir('/MFMDatasets/MFM_bopf/src/multi_predict/RF/output')
#os.chdir('/MFMDatasets/MFM_bopf/src/multi_predict/SVC/output')
#os.chdir('/MFMDatasets/MFM_bopf/src/multi_predict/LSVC/output')
#os.chdir('/MFMDatasets/MFM_bopf/src/multi_predict/MLP/output')
results_df = pd.DataFrame()
#files = glob.glob("*-NB-*.csv")
files = glob.glob("*-LR-*.csv")
#files = glob.glob("*-RF-*.csv")
#files = glob.glob("*-SVC-*.csv")
#files = glob.glob("*-LSVC-*.csv")
#files = glob.glob("*-MLP-*.csv")
files.sort(key=os.path.getmtime)
for idx, f in enumerate(files):
    if results_df.empty:
        results_df = pd.read_csv(f, header=None, index_col=0, names=[idx])
    else:
        this_df = pd.read_csv(f, header=None, index_col=0, names=[idx])
        results_df = results_df.merge(this_df, left_index=True, right_index=True, how="right")
results_df
```

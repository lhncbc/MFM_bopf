```
colnames = preg_df.columns.to_list()
lower_list = [item.lower() for item in colnames]
for item in lower_list:
    if item.find('weight') >= 0 or \
       item.find('bmi') >= 0 or \
       item.find('fhr') >= 0 or \
       item.find('temp') >= 0 or \
       item.find('sbp') >= 0 or \
       item.find('dbp') >= 0 or \
       item.find('age') >= 0:
        print(f'{item}')
```

````python
# Dec.19 - filtering by onepreg
continuous_cols = ['MomID', 'Momage', 'Preweight', 'Admweight', 'Birthweight1', 
                   'Admtemp', 'AdmSBP', 'AdmDBP', 'AdmFHR', 'onepregflag']
preg_df = pd.read_csv('../data/csl/preg_link.csv', usecols=continuous_cols)
#preg_df = pd.read_csv('../data/csl/preg_mwb.csv')
preg_df
````

````python
for col in preg_df:
    nulls = preg_df[col].isnull().sum()
    print(f'{col}: {nulls}')
````

````python
# Using median to impute NULL values. Not sure about statistical validity, especially for
# fields that have large numbers of NULLs
preg_mod = preg_df.fillna(value=preg_df.median(), inplace=False)
preg_mod
````

````python
merged = pd.merge(csl_df, preg_mod, how='inner', left_on='MomID', right_on='MomID')
````

````python
for col in merged:
    if merged[col].isna().any():
        print(col)
````

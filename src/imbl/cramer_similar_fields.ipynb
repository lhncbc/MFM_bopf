```python
ty_all_df = pd.read_csv('/slurm_storage/mbopf/projects/MFM_bopf/src/imbl/multiCorr_2020_08/underCompOut/2020_08/FULLcorr-transfus_yes-NONE-1.0-Cramer-1-20200824_181721.csv', index_col=0)
tl_all_df =  pd.read_csv('/slurm_storage/mbopf/projects/MFM_bopf/src/imbl/multiCorr_2020_08/underCompOut/2020_08/FULLcorr-trans_loss-NONE-1.0-Cramer-1-20200828_114435.csv', index_col=0)
```

```python
abrupt_list = []
for col in ty_all_df:
    if 'brupt' in col:
        abrupt_list.append(col)
        
abrupt_df = ty_all_df.loc[abrupt_list, abrupt_list]
abrupt_df
```

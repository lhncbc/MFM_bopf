```python
import numpy as np
import pandas as pd
import os, glob
from IPython.core.display import display, HTML
pd.options.display.float_format = '{:0.4f}'.format
```

# Cramer RAND: create top50 & union50
### Transfus_yes and trans_loss

```python
basedir = '/slurm_storage/mbopf/projects/MFM_bopf/src/imbl/underCompOut/'
for f in glob.glob(basedir + 'corr-transfus_yes-RAND-1.0-Cramer*.csv'):
    ty_corr_RAND_C = pd.read_csv(f, index_col=0)
for f in glob.glob(basedir + 'rank-transfus_yes-RAND-1.0-Cramer*.csv'):
    ty_rank_RAND_C = pd.read_csv(f, index_col=0)
for f in glob.glob(basedir + 'corr-trans_loss-RAND-1.0-Cramer*.csv'):
    tl_corr_RAND_C = pd.read_csv(f, index_col=0)
for f in glob.glob(basedir + 'rank-trans_loss-RAND-1.0-Cramer*.csv'):
    tl_rank_RAND_C = pd.read_csv(f, index_col=0)
```

```python
pd.options.display.max_rows = None
combined = pd.DataFrame(ty_corr_RAND_C['mean_corr']).rename(columns={'mean_corr':"ty_corr"})
next_df = pd.DataFrame(ty_rank_RAND_C['mean_rank']).round(0).astype(int)
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_rank':"ty_rank"}, inplace=True)
```

```python
next_df = pd.DataFrame(tl_corr_RAND_C['mean_corr']).rename(columns={'mean_corr':"tl_corr"})
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
next_df = pd.DataFrame(tl_rank_RAND_C['mean_rank']).round(0).astype(int)
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_rank':"tl_rank"}, inplace=True)
display(combined.head(15))
len(combined)
```

```python
np.set_printoptions(precision=4) # Limits precision - Does this do anything in Pandas?
out_flag = True
```

```python
# Prevent accidental overwrite by checking if exists; comment out "if" to overwrite
filename = '/MFMDatasets/MFM_bopf/data/csl/CramerTheil/Cramer_PI_TyTl_coeff_rank_FULL.csv'
if out_flag:
    combined.to_csv(filename, sep="\t")
```

```python
filename = '/MFMDatasets/MFM_bopf/data/csl/CramerTheil/Cramer_PI_Ty_coeff_top50.csv'
if out_flag:
    combined['ty_corr'][0:50].to_csv(filename, sep="\t", header=False)
    
filename = '/MFMDatasets/MFM_bopf/data/csl/CramerTheil/Cramer_PI_Ty_vars_top50.csv'
if out_flag:
    combined['ty_corr'][0:50].to_csv(filename, columns=[], header=False )
```

# Trans_loss sort

```python
tl_sort = combined.sort_values(by='tl_rank',ascending=True)
tl_sort
```

```python
filename = '/MFMDatasets/MFM_bopf/data/csl/CramerTheil/Cramer_PI_Tl_coeff_top50.csv'
if out_flag:
    tl_sort['tl_corr'][0:50].to_csv(filename, sep="\t", header=False)
filename = '/MFMDatasets/MFM_bopf/data/csl/CramerTheil/Cramer_PI_Tl_vars_top50.csv'
if out_flag:
    tl_sort['ty_corr'][0:50].to_csv(filename, columns=[], header=False )
```

# Union50 - transfus_yes sort

```python
union50_ty_tl = combined[(combined['ty_rank'] <= 51) | (combined['tl_rank'] <= 51)]
print(len(union50_ty_tl))
display(union50_ty_tl)
```

```python
filename = '/MFMDatasets/MFM_bopf/data/csl/CramerTheil/Cramer_PI_TyTl_coeff_rank_Union50.csv'
if out_flag:
    union50_ty_tl.to_csv(filename, sep="\t")
filename = '/MFMDatasets/MFM_bopf/data/csl/CramerTheil/Cramer_PI_Ty_coeff_Union50.csv'
if out_flag:
    union50_ty_tl['ty_corr'].to_csv(filename, sep="\t", header=False)
filename = '/MFMDatasets/MFM_bopf/data/csl/CramerTheil/Cramer_PI_Ty_vars_Union50.csv'
if out_flag:
    union50_ty_tl['ty_corr'].to_csv(filename, columns=[], header=False )
```

# Union50 - trans_loss sort

```python
union50_tl_sort = union50_ty_tl.sort_values(by='tl_rank',ascending=True)
union50_tl_sort
```

```python
filename = '/MFMDatasets/MFM_bopf/data/csl/CramerTheil/Cramer_PI_Tl_coeff_Union50.csv'
if out_flag:
    union50_tl_sort['tl_corr'].to_csv(filename, sep="\t", header=False)
filename = '/MFMDatasets/MFM_bopf/data/csl/CramerTheil/Cramer_PI_Tl_vars_Union50.csv'
if out_flag:
    union50_tl_sort['tl_corr'].to_csv(filename, columns=[], header=False )
```

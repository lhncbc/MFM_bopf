```python
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
from sklearn.preprocessing import LabelEncoder
def encode_df(df):
    df_enc = pd.DataFrame()
    for col in df:
        le = LabelEncoder()
        le.fit(df[col])
        df_enc[col] = le.transform(df[col])
    return df_enc
```

```python
# Given input and output filenames, LabelEncode input
def encode_file (infile, outfile):
    df = pd.read_csv(infile , header=0, index_col=0)
    df_enc = encode_df(df)
    df_enc.to_csv(outfile)
```

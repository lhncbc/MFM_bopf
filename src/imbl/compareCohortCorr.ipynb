```python
combined = pd.DataFrame(ty_host_T['mean_corr']).rename(columns={'mean_corr':"Host_T"})
combined_C = pd.DataFrame(ty_host_C['mean_corr']).rename(columns={'mean_corr':"Host_C"})
combined
```

```python
next_df = pd.DataFrame(ty_Ins_T['mean_corr'])
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_corr':"Ins_T"}, inplace=True)
combined
```

```python
next_df = pd.DataFrame(ty_Ins_C['mean_corr'])
combined_C = pd.merge(combined_C, next_df, left_index=True, right_index=True)
combined_C.rename(columns={'mean_corr':"Ins_C"}, inplace=True)
combined_C
```

```python
next_df = pd.DataFrame(ty_age_T['mean_corr'])
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_corr':"Age_T"}, inplace=True)
combined
```

```python
next_df = pd.DataFrame(ty_age_C['mean_corr'])
combined_C = pd.merge(combined_C, next_df, left_index=True, right_index=True)
combined_C.rename(columns={'mean_corr':"Age_C"}, inplace=True)
combined_C
```

```python
next_df = pd.DataFrame(ty_momrace_T['mean_corr'])
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_corr':"race_T"}, inplace=True)
combined.head(20)
```

```python
next_df = pd.DataFrame(ty_momrace_C['mean_corr'])
combined_C = pd.merge(combined_C, next_df, left_index=True, right_index=True)
combined_C.rename(columns={'mean_corr':"race_C"}, inplace=True)
combined_C.head(20)
```

```python
next_df = pd.DataFrame(ty_RAND_T['mean_corr'])
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_corr':"RAND_T"}, inplace=True)
combined.head(20)
```

```python
next_df = pd.DataFrame(ty_RAND_C['mean_corr'])
combined_C = pd.merge(combined_C, next_df, left_index=True, right_index=True)
combined_C.rename(columns={'mean_corr':"RAND_C"}, inplace=True)
combined_C.head(20)
```

```python
# Add column for the average rank and sort by it
combined['mean_corr'] = combined.apply(np.average, axis=1)
combined.sort_values(by='mean_corr', ascending=False, inplace=True)
combined.head(20)
```

```python
# Add column for the average rank and sort by it
combined_C['mean_corr'] = combined_C.apply(np.average, axis=1)
combined_C.sort_values(by='mean_corr', ascending=False, inplace=True)
combined_C.head(20)
```

```python
import numpy as np
import pandas as pd
import os, glob
pd.options.display.float_format = '{:0.2f}'.format
```

```python
# @TODO: Change hard-coded paths to wildcards
t_yes_rank = pd.read_csv('/MFMDatasets/MFM_bopf/data/csl/Final_cohort_rank-transfus_yes-Cramer.csv', index_col=0)
t_loss_rank = pd.read_csv('/MFMDatasets/MFM_bopf/data/csl/Final_cohort_rank-trans_loss-Cramer.csv', index_col=0)
t_all_rank = pd.read_csv('/MFMDatasets/MFM_bopf/data/csl/Final_cohort_rank-transfus_all-Cramer.csv', index_col=0)
```

```python
combined = pd.DataFrame(t_loss_rank['mean_rank'].rank().astype('int')).rename(columns={'mean_rank':"trans_loss"})
combined
```

```python
next_df = pd.DataFrame(t_all_rank['mean_rank'].rank().astype('int'))
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_rank':"transfus_all"}, inplace=True)
combined
```

```python
next_df = pd.DataFrame(t_yes_rank['mean_rank'].rank().astype('int'))
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_rank':"transfus_yes"}, inplace=True)
combined
```

```python
# Add column for the average rank and sort by it
combined['mean'] = combined.apply(np.average, axis=1)
combined['mean_rank'] = combined.apply(np.average, axis=1).rank().astype('int')
combined.sort_values(by='mean_rank', ascending=True, inplace=True)
combined.head(50)
```

```python
# NOTE - saving this in /MFMDatasets instead of /slurm_storage
filename = '/MFMDatasets/MFM_bopf/data/csl/Final_Dataset_rank-Cramer.csv'
if not os.path.exists(filename):
    combined.to_csv(filename)
```

```python
import numpy as np
import pandas as pd
import os, glob
pd.options.display.float_format = '{:0.2f}'.format
```

```python
# @TODO: Change hard-coded paths to wildcards
transfus_yes_rank = pd.read_csv('/MFMDatasets/MFM_bopf/data/csl/Final_cohort_rank-transfus_yes-Cramer.csv', index_col=0)
trans_loss_rank = pd.read_csv('/MFMDatasets/MFM_bopf/data/csl/Final_cohort_rank-trans_loss-Cramer.csv', index_col=0)
trans_hyster_rank = pd.read_csv('/MFMDatasets/MFM_bopf/data/csl/Final_cohort_rank-transfus_hyster-Cramer.csv', index_col=0)
hysterectomy_rank = pd.read_csv('/MFMDatasets/MFM_bopf/data/csl/Final_cohort_rank-Hysterectomy-Cramer.csv', index_col=0)

# No longer used
#t_all_rank = pd.read_csv('/MFMDatasets/MFM_bopf/data/csl/Final_cohort_rank-transfus_all-Cramer.csv', index_col=0)
```

```python
transfus_yes_rank.head(200)
```

```python
combined = pd.DataFrame(transfus_yes_rank['mean_rank'].rank().astype('int')).rename(columns={'mean_rank':"transfus_yes"})
combined
```

```python
next_df = pd.DataFrame(trans_hyster_rank['mean_rank'].rank().astype('int'))
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_rank':"trans_hyster"}, inplace=True)
combined
```

```python
next_df = pd.DataFrame(trans_loss_rank['mean_rank'].rank().astype('int'))
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_rank':"trans_loss"}, inplace=True)
combined
```

```python
next_df = pd.DataFrame(hysterectomy_rank['mean_rank'].rank().astype('int'))
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_rank':"Hysterectomy"}, inplace=True)
combined
```

```python
# Add column for the average rank and sort by it
combined['mean_rank'] = combined.apply(np.average, axis=1)
#combined['mean_rank'] = combined.apply(np.average, axis=1).rank().astype('int')
combined.sort_values(by='mean_rank', ascending=True, inplace=True)
combined.head(50)
```

```python
# NOTE - saving this in /MFMDatasets instead of /slurm_storage
filename = '/MFMDatasets/MFM_bopf/data/csl/Final_Dataset_rank-Cramer.csv'
if not os.path.exists(filename):
    combined.to_csv(filename)
```

```python
#combined.drop('mean', axis=1, inplace=True)
combined.drop('mean_rank', axis=1, inplace=True)
combined.head()
```

```python
# Recalc average without Hysterectomy
combined['mean3'] = combined[['transfus_yes','trans_hyster','trans_loss']].apply(np.average, axis=1)
#combined['mean_rank'] = combined.apply(np.average, axis=1).rank().astype('int')
#combined.sort_values(by='mean_rank', ascending=True, inplace=True)
combined.sort_values(by='mean3', ascending=True, inplace=True)
combined.head(50)
```

```python
# NOTE - saving this in /MFMDatasets instead of /slurm_storage
filename = '/MFMDatasets/MFM_bopf/data/csl/Final_Dataset_mean3_rank-Cramer.csv'
if not os.path.exists(filename):
    combined.to_csv(filename)
```

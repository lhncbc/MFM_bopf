```python
ty1 = csl[csl['transfus_yes'] == 1]
#ty1_age_dist = ty1['high_Age'].value_counts(normalize=True, sort=False)
#ty1_dist = ty1['momrace_new'].value_counts(normalize=True, sort=False).sort_index()
#ty1_dist = ty1['Insurance'].value_counts(normalize=True, sort=False).sort_index()
ty1_dist = ty1['Hostype'].value_counts(normalize=True, sort=False).sort_index()
ty1_dist
```

```python
ty2 = csl[csl['transfus_yes'] == 2]
#ty2_dist = ty2['momrace_new'].value_counts(normalize=True, sort=False).sort_index()
#ty2_dist = ty2['Insurance'].value_counts(normalize=True, sort=False).sort_index()
ty2_dist = ty2['Hostype'].value_counts(normalize=True, sort=False).sort_index()
ty2_dist
```

```python
samp_dist = ty2_dist / ty1_dist
samp_dist
```

```python
#var_dist = ty1['high_Age'].map(samp_dist)
#var_dist = ty1['momrace_new'].map(samp_dist)
#var_dist = ty1['Insurance'].map(samp_dist)
var_dist = ty1['Hostype'].map(samp_dist)
var_dist
```

```python
samp0 = ty1.sample(n=5170, weights=var_dist, random_state=0)
#print(samp0['high_Age'].value_counts(normalize=True, sort = False))
#print(samp0['momrace_new'].value_counts(normalize=True, sort = False).sort_index())
#print(samp0['Insurance'].value_counts(normalize=True, sort = False).sort_index())
print(samp0['Hostype'].value_counts(normalize=True, sort = False).sort_index())
samp1 = ty1.sample(n=5170, weights=var_dist, random_state=1)
#print(samp1['high_Age'].value_counts(normalize=True, sort = False))
#print(samp1['momrace_new'].value_counts(normalize=True, sort = False).sort_index())
#print(samp1['Insurance'].value_counts(normalize=True, sort = False).sort_index())
print(samp1['Hostype'].value_counts(normalize=True, sort = False).sort_index())
```

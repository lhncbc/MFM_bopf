```python
import numpy as np
import pandas as pd
import os, glob
pd.options.display.float_format = '{:0.2f}'.format
```

```python
# @TODO: Change hard-coded paths to wildcards
#dataset = 'transfus_yes'
#dataset = 'trans_loss'
#dataset = 'transfus_hyster'
dataset = 'Hysterectomy'
for f in glob.glob('/slurm_storage/mbopf/projects/MFM_bopf/src/imbl/underCompOut/rank-' + dataset + '-Hostype-1.0-Cramer*.csv'):
    ty_host_C = pd.read_csv(f, index_col=0)
for f in glob.glob('/slurm_storage/mbopf/projects/MFM_bopf/src/imbl/underCompOut/rank-' + dataset + '-Insurance-1.0-Cramer*.csv'):
    ty_Ins_C = pd.read_csv(f, index_col=0)
for f in glob.glob('/slurm_storage/mbopf/projects/MFM_bopf/src/imbl/underCompOut/rank-' + dataset + '-high_Age-1.0-Cramer*.csv'):
    ty_age_C = pd.read_csv(f, index_col=0)
for f in glob.glob('/slurm_storage/mbopf/projects/MFM_bopf/src/imbl/underCompOut/rank-' + dataset + '-momrace_new-1.0-Cramer*.csv'):
    ty_momrace_C = pd.read_csv(f, index_col=0)
for f in glob.glob('/slurm_storage/mbopf/projects/MFM_bopf/src/imbl/underCompOut/rank-' + dataset + '-RAND-1.0-Cramer*.csv'):
    ty_RAND_C = pd.read_csv(f, index_col=0)
    
# Not currently used
#dataset = 'transfus_all'
```

```python
combined = pd.DataFrame(ty_host_C['mean_rank'].rank().astype('int')).rename(columns={'mean_rank':"Host_C"})
combined.head(20)
```

```python
next_df = pd.DataFrame(ty_Ins_C['mean_rank'].rank().astype('int'))
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_rank':"Ins_C"}, inplace=True)
combined.head(20)
```

```python
next_df = pd.DataFrame(ty_age_C['mean_rank'].rank().astype('int'))
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_rank':"Age_C"}, inplace=True)
combined.head(20)
```

```python
next_df = pd.DataFrame(ty_momrace_C['mean_rank'].rank().astype('int'))
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_rank':"race_C"}, inplace=True)
combined.head(20)
```

```python
next_df = pd.DataFrame(ty_RAND_C['mean_rank'].rank().astype('int'))
combined = pd.merge(combined, next_df, left_index=True, right_index=True)
combined.rename(columns={'mean_rank':"RAND_C"}, inplace=True)
combined.head(30)
```

```python
# Remove transfus_yes, Hostype, Insurance, high_Age and momrace_new from DataFrame
#combined.drop(['transfus_yes','Hostype', 'Insurance', 'high_Age', 'momrace_new'], inplace=True)
#combined.drop(['trans_loss','Hostype', 'Insurance', 'high_Age', 'momrace_new'], inplace=True)
#combined.drop(['transfus_hyster','Hostype', 'Insurance', 'high_Age', 'momrace_new'], inplace=True)
combined.drop(['Hysterectomy','Hostype', 'Insurance', 'high_Age', 'momrace_new'], inplace=True)

# transfus_all no longer a target
#combined.drop(['transfus_all','Hostype', 'Insurance', 'high_Age', 'momrace_new'], inplace=True)
```

```python
# Add column for the average rank and sort by it
combined['mean_rank'] = combined.apply(np.average, axis=1).rank()
combined.sort_values(by='mean_rank', ascending=True, inplace=True)
combined.head(50)
```

```python
pd.options.display.max_columns = 30
# Note: Name currently assumes updated symbolic link
os.chdir('/MFMDatasets/MFM_bopf/data/csl')
tl_PI = pd.read_csv('CSL_tl_PI_comp.csv', index_col=0)
tl_PI
```

```python
tl_PI['Sitenum'].value_counts(sort=False, dropna=False)
```

```python
for col in tl_PI:
    print(f'type({col}) = {type(tl_PI[col][0])}')
```

```python
def write_site(full_df, sitenum, basename):
    filename = basename + '_s' + str(sitenum) + '.csv'
    print(filename)
    site_df = full_df[full_df['Sitenum'] == sitenum]
    site_df.to_csv(filename, header=True)
```

```python
outflag = False

for site in tl_PI['Sitenum'].unique():
    if outflag:
        write_site(tl_PI, site, 'Sites/CSL_tl_PI')
    else:
        print(f'Site {site} not written out')
```

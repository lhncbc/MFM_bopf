## CSL_NearMiss 
Create "MomNearMiss" field by combining "Momdeath" and "MomICU" and moving in the front.
Save datafile for future use.

```python
import sys
import pandas as pd
import util_mwb
print(sys.version)
print(pd.__version__) # Must be 0.24.1 for IntegerArray
```

```python
%%time
csl_lr = util_mwb.read_csl_csv('../data/csl/MFM_data1.csv')
```

```python
pd.options.display.max_columns = 20
csl_lr
```

```python
print(csl_lr['Momdeath'].value_counts())
print(csl_lr['MomICU'].value_counts())
csl_lr.loc[csl_lr['Momdeath'] == 2]
```

```python
# Calculate 'MomNearMiss' by merging 'Momdeath' and 'MomICU'
csl_lr.loc[(csl_lr['Momdeath'] == 2) | (csl_lr['MomICU'] == 2 ), 'MomNearMiss'] = 2
csl_lr['MomNearMiss'].fillna(1, inplace=True)
csl_lr['MomNearMiss'] = csl_lr['MomNearMiss'].astype('Int8')
csl_lr.shape
```

```python
csl_lr['MomNearMiss'].value_counts()
```

```python
col_list = list(csl_lr.columns)
new_list = [x for x in col_list if x not in ['MomNearMiss', 'Momdeath', 'MomICU']]
new_list.insert(2, 'MomICU')
new_list.insert(2, 'Momdeath')
new_list.insert(2, 'MomNearMiss')
df = csl_lr[new_list]
```

```python
len(new_list)
```

```python
print(df.shape)
df.head()
```

```python
df.iloc[:,0].dtype
```

```python
import numpy as np
df.loc[:, df.dtypes == np.object]
```

```python
# Removed extraneous characters from strings; had to use .loc[row, col] to avoid warning
df.loc[:,'MomID'] = df['MomID'].str.replace('b','')
df.loc[:,'MomID'] = df['MomID'].str.replace("'","")
df.loc[:,'Sitenum'] = df['Sitenum'].str.replace('b','')
df.loc[:,'Sitenum'] = df['Sitenum'].str.replace("'","")
df.head()
```

```python
df.to_csv('../data/csl/CSL_NearMiss.csv')
```

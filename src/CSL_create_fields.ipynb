###  Need to fix some edge conditions in "high_height"

```python
csl['high_height'].value_counts()
```

```python
csl.loc[(csl['high_height'] > 1.00) & (csl['high_height'] < 1.50), 'high_height'] = 1
csl.loc[(csl['high_height'] > 1.50) & (csl['high_height'] < 2.0), 'high_height'] = 2
csl['high_height'] = csl['high_height'].astype(int)
csl['high_height'].value_counts()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
np.set_printoptions(suppress=True)
csl_norm = csl.copy()
col_names = ['EBLoss', 'Admcontract', 'AdmDBP', 'Admefface', 'AdmSBP', 'BESTGA']
features = csl_norm[col_names]
# NOTE: might want to consider Robust or Standard scaler for EBLoss because of outlyers.
#scaler = StandardScaler()
#scaler = RobustScaler()
scaler = MinMaxScaler(feature_range=(-1,1))
scaler.fit(features.values)
features = scaler.transform(features.values)
csl_norm[col_names] = features
for col in col_names:
    print(col)
    print(csl[col].value_counts().index.values)
    # Limit precision of scaled variables to 4 decimal places
    csl_norm[col] = csl_norm[col].apply(lambda x:np.round(x, decimals=4))
    print(csl_norm[col].value_counts().index.values)
    
```

```python
from datetime import datetime
datestr = datetime.today().strftime('%Y-%m-%d')

out_flag = False
```

```python
# Write out both csl & csl_mod

# "CSL_d6_fields" contains d6 fields with all possible targets included.
# It is intended to be subsetted since many of the targest and variables overlap and 
# would cause multicollinearity if this file were used in predictions.
filename = 'MFM_CSL_d6_fields_' + datestr + '.csv'
linkname = 'CSL_d6_fields.csv'
# This version is not normalized
if out_flag:
    csl.to_csv(filename)
    os.remove(linkname)
    os.symlink(filename, linkname)
else:
    print("out_flag False: File not written")
    
# This version is has been scaled/normalized 
norm_file = 'MFM_CSL_d6_fields_norm_' + datestr + '.csv'
linkname = 'CSL_d6_fields_norm.csv'
if out_flag:
    csl_norm.to_csv(norm_file)
    os.remove(linkname)
    os.symlink(norm_file, linkname)
else:
    print("out_flag False: norm_file not written")
```

# Convert Sitenum to int

```python
print(type(csl_df['Sitenum']))
csl_df['Sitenum'] = csl_df['Sitenum'].astype(int)
print(type(csl_df['Sitenum']))
```

```python
csl_df['Momage'].value_counts(sort=False, dropna=False).sort_index()
```

```python
csl_df['PreBMI'].value_counts(sort=False, dropna=False).sort_index()
```

# Impute NaN to median

```python
bmi_median = csl_df['PreBMI'].median()
age_median = csl_df['Momage'].median()
print(f'bmi_median = {bmi_median}')
print(f'age_median = {age_median}')
csl_df['new_BMI'] = csl_df['PreBMI'].fillna(bmi_median)
csl_df['new_age'] = csl_df['Momage'].fillna(age_median)
print(csl_df['new_BMI'].value_counts(dropna=False))
print(csl_df['new_age'].value_counts(dropna=False))
```

# Round new_BMI 

```python
csl_df['new_BMI'] = csl_df['new_BMI'].round().astype(int)
csl_df['new_BMI'].value_counts(dropna=False)
```

```python
tl_PI_mod = tl_PI.merge(csl_df[['new_age', 'new_BMI']], left_index=True, right_index=True, how='inner')
tl_PI_mod
```

# Create two versions: high_age_bmi (ordinal) and age_bmi (continuous)

```python
tl_PI_mod.drop(['high_Age', 'high_BMI'], axis=1, inplace=True)
high_age_bmi = tl_PI_mod.copy()
high_age_bmi = high_age_bmi.drop(['new_age', 'new_BMI'], axis=1)
age_bmi_new = tl_PI_mod.drop(['new_high_Age', 'new_high_BMI'], axis=1)
display(high_age_bmi)
display(age_bmi_new)
display(tl_PI_mod)
```

### Use "out_flag" variable to control whether files are written or not

```python
out_flag = False

from datetime import datetime
datestr = datetime.today().strftime('%Y-%m-%d')
```

```python
# Write out trans_loss, preIntra, both continuous and ordinal age, bmi
filename = 'MFM_CSL_d6_tl_PI_age-bmi_both' + datestr + '.csv'
linkname = 'CSL_tl_PI_ab-both.csv'
if out_flag:
    tl_PI_mod.to_csv(filename, header=True)
    if os.path.exists(linkname):
        os.remove(linkname)
    os.symlink(filename, linkname)
else:
    print(f'File not written: {filename}')
```

```python
# Write out trans_loss, preIntra, continuous age, bmi
filename = 'MFM_CSL_d6_tl_PI_age-bmi' + datestr + '.csv'
linkname = 'CSL_tl_PI_ab-cont.csv'
if out_flag:
    age_bmi_new.to_csv(filename, header=True)
    if os.path.exists(linkname):
        os.remove(linkname)
    os.symlink(filename, linkname)
else:
    print(f'File not written: {filename}')
```

```python
# Write out trans_loss, preIntra, ordinal high_Age, high_BMI
filename = 'MFM_CSL_d6_tl_PI_high_age-bmi' + datestr + '.csv'
linkname = 'CSL_tl_PI_high-ab.csv'
if out_flag:
    high_age_bmi.to_csv(filename, header=True)
    if os.path.exists(linkname):
        os.remove(linkname)
    os.symlink(filename, linkname)
else:
    print(f'File not written: {filename}')
```

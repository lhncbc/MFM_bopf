# Read "data5" 

```python
os.chdir('/MFMDatasets/MFM_bopf/data/csl')
d5 = pd.read_csv('MFM_data5.csv', index_col=0)
d5
```

```python
for col in d5:
    if d5[col].value_counts().index.size == 1:
        print(f'{col} has only 1 value: NEED to DROP')
        d5.drop(col, inplace=True, axis=1)
```

```python
# Remove extraneous characters from strings; had to use .loc[row, col] to avoid warning
d5.loc[:,'MomID'] = d5['MomID'].str.replace('b','')
d5.loc[:,'MomID'] = d5['MomID'].str.replace("'","")
d5.loc[:,'Sitenum'] = d5['Sitenum'].str.replace('b','')
d5.loc[:,'Sitenum'] = d5['Sitenum'].str.replace("'","")
```

```python
d5.set_index('MomID', inplace=True)
d5.head()
```

```python
pd.options.display.float_format = "{:,.2f}".format
del5_df = pd.DataFrame(d5['Delmode'].value_counts(sort=False)).sort_index()
del5_df['pct'] = pd.DataFrame(d5['Delmode'].value_counts(sort=False, normalize=True))*100
del5_df
```

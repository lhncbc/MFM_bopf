# Create data file for high_EBLoss target 

```python
he_df = pd.read_csv('/MFMDatasets/MFM_bopf/data/csl/high_EBL.txt', index_col=0)
he_df
```

```python
he_tytl_PI = he_df.merge(tytl_PI, left_index=True, right_index=True, how='right')
he_tytl_PI 
```

```python
he_tytl_PI = he_tytl_PI[he_tytl_PI['high_EBL'] < 9]
he_tytl_PI
```

```python
he_ty_PI = he_tytl_PI.drop('trans_loss', axis=1)
he_ty_PI
```

```python
he_ty_PI['high_EBL'].value_counts()
```

```python
out_flag = False

from datetime import datetime
datestr = datetime.today().strftime('%Y-%m-%d')
# Write out complete dataset, including Sitenum
filename = 'MFM_CSL_d6_hety_PI_' + datestr + '.csv'
linkname = 'CSL_hety_PI_comp.csv'
if out_flag:
    he_ty_PI.to_csv(filename, header=True)
    if os.path.exists(linkname):
        os.remove(linkname)
    os.symlink(filename, linkname)
else:
    print(f'File not written: {filename}')
```

```python
he_PI = he_ty_PI.drop('transfus_yes', axis=1)
he_PI
```

```python
out_flag = False
# Write out complete dataset, including Sitenum
filename = 'MFM_CSL_d6_highEBL_PI_' + datestr + '.csv'
linkname = 'CSL_he_PI_comp.csv'
if out_flag:
    he_PI.to_csv(filename, header=True)
    if os.path.exists(linkname):
        os.remove(linkname)
    os.symlink(filename, linkname)
else:
    print(f'File not written: {filename}')
```

```python
total = len(csl)
csl_freq = pd.DataFrame()
for col in csl:
    if col != 'trans_loss':
        var_freq = csl.groupby(col).size()/total
        csl_freq[col] = csl[col].map(var_freq)
    else:
        csl_freq[col] = csl[col]
csl_freq = csl_freq.round(4)
```

```python
csl_freq
```

```python
for col in csl_freq:
    print(f'{col}:\n{csl_freq[col].value_counts(sort=False).sort_index().to_string()}\n')
```

```python
out_flag = False
#out_flag = True

from datetime import datetime
datestr = datetime.today().strftime('%Y-%m-%d')
```

```python
# Write out trans_loss for "pre_intra"
filename = 'MFM_CSL_d6_trans_loss_Pre-Intra_FREQ_' + datestr + '.csv'
linkname = 'CSL_tl_PI_Freq.csv'
if out_flag:
    csl_freq.to_csv(filename, header=True)
    if os.path.exists(linkname):
        os.remove(linkname)
    os.symlink(filename, linkname)
else:
    print(f'File not written: {filename}')
```

### Create new dataset including high_EBL to compare with Venkatesh
Also need to remove all NaNs (which = 9)

```python
high_EBL = pd.read_csv('/MFMDatasets/MFM_bopf/data/csl/high_EBL.txt', index_col=0)
high_EBL
```

```python
merged = csl_freq.merge(high_EBL, left_index=True, right_index=True, how="left")
merged
```

```python
ebl_df = merged[merged['high_EBL'] < 9]
ebl_df
```

```python
ebl_df = ebl_df.drop('trans_loss', axis=1)
```

```python
ebl_df['high_EBL'].value_counts()
```

```python
# Move EBL to front
ebl_col = ebl_df['high_EBL']
ebl_df = ebl_df.drop('high_EBL', axis=1)
ebl_df.insert(0, 'high_EBL', ebl_col)
ebl_df
```

```python
ebl_df['high_EBL'].value_counts()
```

```python
# Write out trans_loss for "pre_intra"
filename = '/MFMDatasets/MFM_bopf/data/csl/CSL_he_PI_Freq.csv'
if out_flag:
    ebl_df.to_csv(filename, header=True)
else:
    print(f'File not written: {filename}')
```

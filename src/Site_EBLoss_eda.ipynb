```python
MFM_typedict = np.load('/MFMDatasets/MFM_bopf/data/csl/MFM_linked-by-preg_typedict.npy', allow_pickle=True).item()
MFM_lbp = pd.read_csv('/MFMDatasets/MFM_bopf/data/csl/MFM_linked-by-preg_Int8.csv', index_col=0, dtype=MFM_typedict)
MFM_lbp['Sitenum'] =  MFM_lbp['Sitenum'].astype(np.int8)
MFM_lbp
```

```python
hab_site = high_age_both.merge(MFM_lbp['Sitenum'], left_index=True, right_index=True, how='left')
hab_site
```

```python
full_df = MFM_lbp[['Sitenum','Posttransfus','Bloodproduct','EBLoss']].merge(high_age_both, left_index=True, right_index=True, how='right')
full_df
```

```python
full_df['transfus_yes'] = 0
full_df.loc[(full_df['Posttransfus'] == 2) | (full_df['Bloodproduct'] == 2), 'transfus_yes'] = 1
full_df['transfus_yes'].value_counts()
```

```python
full_df['high_EBL'] = 1
full_df.loc[(full_df['EBLoss'] >= 1000), 'high_EBL'] = 2
full_df.loc[np.isnan(full_df['EBLoss']), 'high_EBL'] = 9
full_df['high_EBL'].value_counts()
```
